```python
df['started_at']=pd.to_datetime(df['started_at'])
df['ended_at']=pd.to_datetime(df['ended_at'])
```

```python
df.dtypes
```

```python

list_data = []

for i in range(len(df)):
    x = tuple(
        None if pd.isna(value) else float(value) if isinstance(value, (float, int, np.float64, np.int64)) else value
        for value in df.iloc[i]
    )
    list_data.append(x)

list_data
```

```python
len(list_data)
```

```python
file = open('processed_trip_data_oct_2023.txt', 'w')

for tuple in list_data:
    line = str(tuple).replace('None', 'null') + ',\n'
    file.write(line)

file.close()
```
